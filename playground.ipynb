```python
from BM25_init import BM25Simple
import pickle
import os

with open("wiki_hard.pkl", "rb") as f:
    data = pickle.load(f)

model = BM25Simple(data)
```

```python
queries_path = os.path.join("test_data", "wiki_hard_test_50-100.pkl")

with open(queries_path, "rb") as f:
    queries_data = pickle.load(f)

queries_data
```

```python
queries = queries_data["text"]

for query in queries:
    # Retrieves top 20
    top_n = model.get_top_n(query)
    with open("test2.txt", "a+") as f:
        f.write("Query:")
        f.write(query)
        f.write("\nTop_n:")
        f.write(str(top_n))
        f.write("\n\n")
```

```python
queries = queries_data["text"]

for i, query in enumerate(queries):

    # Retrieves top 20
    words = query.replace("\xa0", "").replace("\n", "").replace("\\", "").split(" ")
    frequencies = {}
    for word in words:
        if word not in frequencies:
            frequencies[word] = 0
        frequencies[word] += 1
    print({k: v for k, v in sorted(frequencies.items(), key=lambda item: item[1], reverse=True)})

    if i == 50:
        break
    
```

```python
for i in range(70):
    with open("counts.txt", "a") as f:
        f.write("\n\n")
        f.write(model.title[i])
        f.write(", ")
        f.write(str(model.dl[i]))
        f.write("\n\n")
        res = {k: v for k, v in sorted(model.f[i].items(), key=lambda item: item[1], reverse=True)}
        res = {k: v for k, v in res.items() if v > 5}
        f.write(str(res))
```

```python
for i in range(70):
    with open("ratio.txt", "a") as f:
        f.write("\n\n")
        f.write(model.title[i])
        f.write(", the: ")
        f.write(str(model.f[i]["the"] / model.dl[i]))
        f.write(", of: ")
        f.write(str(model.f[i]["of"] / model.dl[i]))
        f.write(", a: ")
        f.write(str(model.f[i]["a"] / model.dl[i]))
        f.write(", and: ")
        f.write(str(model.f[i]["and"] / model.dl[i]))
        f.write("\n\n")
        res = {k: v for k, v in sorted(model.f[i].items(), key=lambda item: item[1], reverse=True)}
        res = {k: v for k, v in res.items() if v > 5}
        f.write(str(res))
```
