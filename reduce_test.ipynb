```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained('facebook/contriever')
model = AutoModel.from_pretrained('facebook/contriever')
```

```python
sentences = [
    "Jajangmyeon is a Chinese-style Korean noodle dish topped with a thick sauce made of chunjang, diced pork, and vegetables.",
    "It originated in Incheon, Korea where Chinese migrant workers started making zhajiangmian (noodles served with fried bean sauce) in the late 19th century.",
    "Modifications in Korea such as a darker and sweeter sauce differentiate the Korean version from the Chinese dish."
]

# apply tokenizer
inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# compute token embeddings
outputs = model(**inputs)
```

```python
# mean pooling
def mean_pooling(token_embeddings, mask):
    token_embeddings = token_embeddings.masked_fill(~mask[..., None].bool(), 0.)
    sentence_embeddings = token_embeddings.sum(dim=1) / mask.sum(dim=1)[..., None]
    return sentence_embeddings

embeddings = mean_pooling(outputs[0], inputs['attention_mask'])
```

```python
embeddings_np = embeddings.detach().numpy().shape
```
